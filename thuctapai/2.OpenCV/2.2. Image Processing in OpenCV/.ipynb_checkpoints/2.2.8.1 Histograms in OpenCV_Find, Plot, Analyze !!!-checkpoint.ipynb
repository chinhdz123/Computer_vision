<h3>Theory<h3>

Vậy biểu đồ là gì? Bạn có thể coi biểu đồ là một biểu đồ hoặc biểu đồ, cho bạn ý tưởng tổng thể về phân phối cường độ của hình ảnh. Đó là một biểu đồ với các giá trị pixel (từ 0 đến 255, đôi lúc không phải) trong trục x và số lượng pixel tương ứng trong hình ảnh trên trục y. Nó chỉ là một cách khác để hiểu hình ảnh. Bằng cách nhìn vào biểu đồ của một hình ảnh, bạn có trực giác về độ tương phản, độ sáng, phân phối cường độ, vv của hình ảnh đó. Hầu như tất cả các công cụ xử lý hình ảnh ngày nay, cung cấp các tính năng trên biểu đồ.

![histogram_sample](histogram_sample.JPG)

Bạn có thể thấy hình ảnh và biểu đồ của nó. (Hãy nhớ rằng, biểu đồ này được vẽ cho hình ảnh thang độ xám, không phải hình ảnh màu). Vùng bên trái của biểu đồ cho thấy lượng pixel tối hơn trong hình ảnh và vùng bên phải cho thấy lượng pixel sáng hơn. Từ biểu đồ, bạn có thể thấy vùng tối hơn vùng sáng hơn và số lượng midtones (giá trị pixel ở tầm trung, nói rằng khoảng 127) là rất ít.

<h3>Find Histogram<h3>

___BINS___: Biểu đồ trên hiển thị số lượng pixel cho mỗi giá trị pixel, tức là từ 0 đến 255. tức là bạn cần 256 giá trị để hiển thị biểu đồ trên. Nhưng hãy xem xét, điều gì sẽ xảy ra nếu bạn không cần tìm số lượng pixel cho tất cả các giá trị pixel một cách riêng biệt, nhưng số lượng pixel trong một khoảng giá trị pixel? Ví dụ, bạn cần tìm số pixel nằm trong khoảng từ 0 đến 15, sau đó 16 đến 31, ..., 240 đến 255. Bạn sẽ chỉ cần 16 giá trị để biểu diễn biểu đồ. Và đó là những gì được thể hiện trong ví dụ được đưa ra trong các hướng dẫn OpenCV về biểu đồ. Vì vậy, những gì bạn làm chỉ đơn giản là chia toàn bộ biểu đồ thành 16 phần phụ và giá trị của mỗi phần phụ là tổng của tất cả số lượng pixel trong đó. Đây mỗi phần phụ được gọi là "thùng". Trong trường hợp đầu tiên, số lượng thùng là 256 (một cho mỗi pixel) trong khi trong trường hợp thứ hai, nó chỉ là 16. BIN được biểu thị bằng thuật ngữ histsize trong tài liệu openCV.

___DIMS:___ Là số lượng tham số mà chúng ta thu thập dữ liệu. Trong trường hợp này, chúng ta thu thập dữ liệu chỉ liên quan đến một thứ, giá trị cường độ. Vì vậy, đây là 1.

___RANGE:___ Là phạm vi giá trị cường độ bạn muốn đo. Thông thường, nó là [0,256], tức là tất cả các giá trị cường độ.

___1. Histogram Calculation in OpenCV___

cv.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])

   * ___images___: nó là hình ảnh nguồn của loại uint8 hoặc float32. nó phải được đặt trong dấu ngoặc vuông, tức là "[img]".
   * ___channels___: nó cũng được cho trong ngoặc vuông. Đây là chỉ số của kênh mà chúng tôi tính toán biểu đồ. Ví dụ: nếu đầu vào là hình ảnh thang độ xám, giá trị của nó là [0]. Đối với hình ảnh màu, bạn có thể vượt qua [0], [1] hoặc [2] để tính toán biểu đồ của kênh màu xanh lam, xanh lục hoặc đỏ tương ứng.
   * ___mask___: hình mặt nạ. Để tìm biểu đồ của hình ảnh đầy đủ, nó được cho là "None". Nhưng nếu bạn muốn tìm biểu đồ của vùng cụ thể của hình ảnh, bạn phải tạo một hình ảnh mặt nạ cho vùng đó và đặt nó dưới dạng mask.
   * ___histSize___: điều này thể hiện số BIN của chúng tôi. Cần phải được cho trong ngoặc vuông. Đối với quy mô đầy đủ, chúng ta sử dụng giá trị [256].
   * ___ranges___ : đây là RANGE của chúng tôi. Thông thường, nó là [0,256].

```python
img = cv.imread('ronaldo.jpg',0)
hist = cv.calcHist([img],[0],None,[256],[0,256])
```

Hist là một mảng 256x1, mỗi giá trị tương ứng với số lượng pixel trong hình ảnh đó với giá trị pixel tương ứng của nó.


___2. Histogram Calculation in Numpy___

Numpy cũng cung cấp cho bạn một hàm, np.histogram (). Vì vậy, thay vì chức năng calchist (), bạn có thể thử bên dưới dòng:

```python
hist,bins = np.histogram(img.ravel(),256,[0,256])
```

<h3>Plotting Histograms<h3>

___1. Using Matplotlib___

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('ronaldo.jpg',0)
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

___Using OpenCV___

Điều chỉnh các giá trị của biểu đồ cùng với các giá trị bin của nó trông giống như tọa độ x, y để bạn có thể vẽ nó bằng hàm cv.line () hoặc cv.polyline () để tạo cùng một hình ảnh như trên. Điều này đã có sẵn với các mẫu chính thức OpenCV-Python2. Kiểm tra mã tại các mẫu/python/hist.py.

<h3>Application of Mask<h3>

```python
img = cv.imread('ronaldo.JPG',0)
# create a mask
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300, 100:400] = 255
masked_img = cv.bitwise_and(img,img,mask = mask)
# Calculate histogram with mask and without mask
# Check third argument for mask
hist_full = cv.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv.calcHist([img],[0],mask,[256],[0,256])
plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```
